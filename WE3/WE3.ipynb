```python
def binary_search(lst, target, left=0, right=None):
    """
    Recursive binary search function.
    
    :param lst: Sorted list of integers
    :param target: Integer value to search for
    :param left: Left boundary of search (default 0)
    :param right: Right boundary of search (default None, which becomes len(lst) - 1)
    :return: Index of target if found, -1 if not found
    """
    if right is None:
        right = len(lst) - 1

    # Base case: target not found
    if left > right:
        return -1

    mid = (left + right) // 2

    if lst[mid] == target:
        return mid
    elif lst[mid] > target:
        # Search in the left half
        return binary_search(lst, target, left, mid - 1)
    else:
        # Search in the right half
        return binary_search(lst, target, mid + 1, right)

# Example usage:
a = [-4, 2, 7, 10, 15, 20, 22, 25, 30, 36, 42, 50, 56, 68, 85, 92, 103]
print(binary_search(a, 42))  # Output: 10
print(binary_search(a, 66))  # Output: -1
```
