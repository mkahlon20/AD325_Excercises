Lemur

```python
def is_looping(n):
    # helper makes next number by squaring each digit and adding
    def next_number(num):
        nxt = 0
        str_num = str(num)          # convert number to iterate digits
        for d in str_num:
            nxt += int(d)**2        # square each digit and add
        return nxt

    visited = set()                 # track seen numbers
    while n != 1:                   # stop if it ever reaches 1
        if n in visited:            # if we saw this before it's looping
            return True
        visited.add(n)              # mark number as seen
        n = next_number(n)          # move to next number
    return False                    # if it hits 1 then no loop
```

Leet

```python
class Solution:
    def minimumOperations(self, nums):
        # set to store unique non zero values
        seen = set()

        for n in nums:
            if n != 0:          # ignore zeros
                seen.add(n)     # track each distinct non zero

        return len(seen)        # answer is how many distinct non zeros
```
