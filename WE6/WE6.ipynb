```python
import heapq

class MedianFinder:

    def __init__(self):
        # small stores the smaller half of the numbers
        # Python only has a min heap, store negatives to simulate a max heap
        self.small = []
        # large will store the larger half of the numbers (normal min heap)
        self.large = []

    def addNum(self, num):
        # push the new number into small
        heapq.heappush(self.small, -num)

        # order between heaps is correct.
        # largest value in small should not be bigger than the smallest in large.
        if self.large and -self.small[0] > self.large[0]:
            # Move the top of small into large to fix the order
            value = -heapq.heappop(self.small)
            heapq.heappush(self.large, value)

        # rebalance the sizes.
        if len(self.small) > len(self.large) + 1:
            # small has too many, so move one to large
            value = -heapq.heappop(self.small)
            heapq.heappush(self.large, value)

        elif len(self.large) > len(self.small) + 1:
            # large has too many, so move one to small
            value = heapq.heappop(self.large)
            heapq.heappush(self.small, -value)

    def findMedian(self):
        # if one heap is bigger, the median is the top of that heap
        if len(self.small) > len(self.large):
            return float(-self.small[0])

        if len(self.large) > len(self.small):
            return float(self.large[0])

        # if heaps are the same size, median is the average of both tops
        return ( (-self.small[0]) + self.large[0] ) / 2.0
```

```python
def max_three(nums):
    # sort the numbers so smallest are first and biggest are last
    nums.sort()

    # multiply the three biggest numbers
    product_big_three = nums[-1] * nums[-2] * nums[-3]

    # multiply the two smallest numbers with the biggest (two negatives can make a big positive)
    product_two_small_one_big = nums[0] * nums[1] * nums[-1]

    # return whichever product is larger
    return max(product_big_three, product_two_small_one_big)
```
