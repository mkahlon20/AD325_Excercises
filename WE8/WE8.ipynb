```python
# Definition for a Node.
# class Node:
#     def __init__(self, val = 0, neighbors = None):
#         self.val = val
#         self.neighbors = neighbors if neighbors is not None else []

class Solution:
    def cloneGraph(self, node):
        if not node:           # empty graph
            return None

        old_to_new = {}        # map original node -> cloned node

        def dfs(n):
            if n in old_to_new:    # already cloned this node
                return old_to_new[n]

            copy = Node(n.val)     # make copy of node
            old_to_new[n] = copy   # store in map

            for nei in n.neighbors:
                copy.neighbors.append(dfs(nei))  # link cloned neighbors

            return copy

        return dfs(node)       # return clone of start node
```
